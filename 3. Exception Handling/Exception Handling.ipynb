# **Exception Handling** (Basic fundemental practice)

*   To write a safe code
*   Because some errors will happen during run time
*   High chance that the show some error at production
*   Try to write all code in exception handling
*   If an error occur at a point of a program, the lines of code blow that will not work.
*   Another posibility is that, the code may run when you are doing it, but in future it may fail.
*   So try to write all code inside a try block. 
*   So that if we write a code inside a try block, if an error occur. Rather than throwing a error, it will throw an exception. 
*   Try block mainly protect our code.
* Only work when your code is perfect but error comes during the exicution. (runtime error)
* Use if you have a doubt an error may occur
* Except block will handle your error.
* If you write a try block you have to write the except block also.



```python
a = 1/0
```

```python
try:
  a = 1/0
except:
  print("error has occured")

#we can see that no error has occured.
```

**try except can be only used for runtime errors, not syntax error, example is shown below**

```python
l = [1, 2, 3]
try:
  for i range(l):
    print(i)
except:
  print("error")
```

```python
l = [1, 2, 3]
try:
  for i in range(len(l)+1):
    print(l[i])
except:
  print("error")
```

**To get what type of error had occured**

```python
l = [1, 2, 3]
try:
  for i in range(len(l)+1):
    print(l[i])
except Exception as e:
  print("error")
  print("error type is ",e)
```

```python
try:
  d = {1 : "j", 2 : "k", 3 : "l"}
  d[4] = int(input())
except Exception as e: #Exception is a superclass, inside contain all the exceptions
  print('ERROR : ', e)
```

Error we are getting is a **Value Error**

Only handle value error

```python
try:
  d = {1 : "j", 2 : "k", 3 : "l"}
  d[4] = int(input())
except ValueError as e: #Exception is a superclass, Value error is a sub class of exception, (100s of subclass)
  print('ERROR : ', e)
```

# **Try Except Else**

Work when the try statement work successfully.

```python
try:
  a = 1/2
except ValueError as e:
  print('ERROR : ', e)
else:
  print("This is the else statement")
```

```python
try:
  a = 1/0


except ValueError as e:
  print('ERROR : ', e)
except ZeroDivisionError as a:
  print("ERRORR : ", a)
except Exception as j:
  print("Super class")
else:
  print("This is the else block")
```

# **Finally**

* Excicute all the time.

```python
try:
  a = 1/0


except ValueError as e:
  print('ERROR : ', e)
except ZeroDivisionError as a:
  print("ERRORR : ", a)
except Exception as j:
  print("Super class")
else:
  print("This is the else block")
finally:
  print("Inside the finally")
```

# **Nested try extend**

When we write the finally code, we can get a doubt like what if the error occured inside the finally block.

The solution is that write a try block inside the finally

```python
try:
  a = 1/0


except ValueError as e:
  print('ERROR : ', e)
except ZeroDivisionError as a:
  print("ERRORR : ", a)
except Exception as j:
  print("Super class")
else:
  print("This is the else block")
finally:
  try:
    b = 1/0
  except:
    print("try inside the finally")
```
